```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("data/bike_sharing.csv")  # Update with your actual file path

# Convert 'dteday' to datetime
df['dteday'] = pd.to_datetime(df['dteday'])

# Extract day, month, year
df['day'] = df['dteday'].dt.day
df['month'] = df['dteday'].dt.month
df['year'] = df['dteday'].dt.year

# Basic info
print(df.info())
print(df.describe())

# Check for nulls
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['cnt'], kde=True, bins=30)
plt.title('Distribution of Total Bike Rentals (cnt)')
plt.xlabel('Total Rentals')
plt.ylabel('Frequency')
plt.show()
```

```python

plt.figure(figsize=(8, 5))
sns.boxplot(x='weathersit', y='cnt', data=df)
plt.title('Bike Rentals by Weather Situation')
plt.xlabel('Weather (1: Clear, 2: Mist, 3: Light Snow/Rain)')
plt.ylabel('Total Rentals')
plt.show()
```

```python


plt.figure(figsize=(8, 5))
sns.boxplot(x='season', y='cnt', data=df)
plt.title('Bike Rentals by Season (1: Spring ... 4: Fall)')
plt.xlabel('Season')
plt.ylabel('Total Rentals')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='month', y='cnt', data=df.groupby('month')['cnt'].mean().reset_index())
plt.title('Average Bike Rentals by Month')
plt.xlabel('Month')
plt.ylabel('Average Rentals')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='weekday', y='cnt', data=df)
plt.title('Total Bike Rentals by Weekday (0: Sunday ... 6: Saturday)')
plt.xlabel('Weekday')
plt.ylabel('Total Rentals')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='temp', y='cnt', data=df, alpha=0.5)
plt.title('Temperature vs Total Rentals')
plt.xlabel('Normalized Temp')
plt.ylabel('Total Rentals')
plt.show()
```
